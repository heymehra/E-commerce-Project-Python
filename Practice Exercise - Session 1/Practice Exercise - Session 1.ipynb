#### Maximum Frequency

Q4:Plot a histogram for the Mins column with 15 bins. Among the three ranges mentioned below, which one has the highest frequency?

A - [54.6,68)

B - [68,81.4)

C - [121.6,135)

D - [1, 14.4)

    a)A - [54.6,68)
    b)B - [68,81.4)
    c)C - [121.6,135)
    d)D - [1, 14.4)


```python
df.head()
```

```python
df.Mins.value_counts()
```

```python
def make_floatt(val):
    val = val.replace('-','')
    return float(val) if val.strip() else 0.0
```

```python
df['Mins'] = df.Mins.apply(make_floatt)
```

```python
#Your code here
plt.figure(figsize=(8, 5))
n, bins, patches = plt.hist(df['Mins'], bins=15, edgecolor='black', alpha=0.7)

# Labels and Title
plt.xlabel('Minutes (Mins)')
plt.ylabel('Frequency')
plt.title('Histogram of Mins')

# Show plot
plt.show()

# Print bins and corresponding frequencies
for i in range(len(bins)-1):
    print(f"Range [{bins[i]:.1f}, {bins[i+1]:.1f}) -> Frequency: {n[i]}")
```
